### 1. How likely is it that you roll doubles when rolling two dice?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

rolls = np.random.choice([1, 2, 3, 4, 5, 6], (10_001, 2))
df = pd.DataFrame(rolls)
df
```

```python
def all_elements_equal(rolls: pd.Series) -> bool:
    return all(x==rolls[0] for x in rolls)
```

```python
df.apply(all_elements_equal, axis=1)
```

```python
df.apply(all_elements_equal, axis=1).mean()
```

### 2. If you flip 8 coins, what is the probability of getting exactly 3 heads? What is the probability of getting more than 3 heads?

```python
# heads = 1; tails = 0
flip_8 = np.random.choice([0, 1], (10_001, 8))
eight_coins_df = pd.DataFrame(flip_8)
eight_coins_df
```

```python
# aggregate by row
n_heads = eight_coins_df.sum(axis=1)
n_heads
```

```python
# prob of getting exactly 3 heads
(n_heads == 3).mean()
```

```python
# prob of getting more than 3 heads
(n_heads > 3).mean()
```

### 3. There are approximitely 3 web development cohorts for every 1 data science cohort at Codeup. Assuming that Codeup randomly selects an alumni to put on a billboard, what are the odds that the two billboards I drive past both have data science students on them?

```python
# web_dev = 0, ds = 1 

billboard = np.random.choice([0, 1], p=[3/4, 1/4], size=(10_001, 2))
billboard_df = pd.DataFrame(billboard)
billboard_df
```

```python
# aggregate by row
ds_studs = billboard_df.sum(axis=1)
ds_studs
```

```python
# prob of both billboards have ds studs
(ds_studs == 2).mean()
```

### 4. Codeup students buy, on average, 3 poptart packages (+- 1.5) a day from the snack vending machine. If on monday the machine is restocked with 17 poptart packages, how likely is it that I will be able to buy some poptarts on Friday afternoon?



```python
mu, sigma = 3, 1.5
weekly_poptarts = np.random.normal(mu, sigma, size=(10_001, 5))
weekly_poptarts = pd.DataFrame(weekly_poptarts)
weekly_poptarts = weekly_poptarts.astype(int)
weekly_poptarts
```

```python
weekly_poptarts['poptarts_sold'] = weekly_poptarts.sum(axis=1)
weekly_poptarts
```

```python
weekly_poptarts['poptarts_left'] = 17 - weekly_poptarts.poptarts_sold
weekly_poptarts
```

```python
weekly_poptarts[weekly_poptarts.poptarts_left < 0] = 0
weekly_poptarts
```

```python
# prob of 'some' poptarts left
(weekly_poptarts.poptarts_left > 1).mean()
```

### 5. Compare Heights

- Men have an average height of 178 cm and standard deviation of 8cm.
- Women have a mean of 170, sd = 6cm.

If a man and woman are chosen at random, P(woman taller than man)?

```python
male_mu, male_sigma = 178, 8
male_heights = np.random.normal(male_mu, male_sigma, size=10_001)

female_mu, female_sigma = 170, 6
female_heights = np.random.normal(female_mu, female_sigma, size=10_001)

heights = pd.DataFrame({'males': male_heights, 'females': female_heights})
heights
```

```python
# prob female is taller than male
(heights.females > heights.males).mean()
```

### 6. When installing anaconda on a student's computer, there's a 1 in 250 chance that the download is corrupted and the installation fails. What are the odds that after having 50 students download anaconda, no one has an installation issue? 100 students?

```python
# install_fail = 0, install_success = 1 

installs = np.random.choice([0, 1], p=[1/250, 249/250], size=(10_001, 50))
installs = pd.DataFrame(installs)
installs
```

```python
installs['successful_installs'] = installs.sum(axis=1)
installs
```

```python
# prob all installs successful
(installs.successful_installs == 50).mean()
```

```python
hundo_installs = np.random.choice([0, 1], p=[1/250, 249/250], size=(10_001, 100))
hundo_installs = pd.DataFrame(hundo_installs)
hundo_installs
```

```python
hundo_installs['successful_installs'] = hundo_installs.sum(axis=1)
hundo_installs
```

```python
# prob all installs successful
(hundo_installs.successful_installs == 100).mean()
```
